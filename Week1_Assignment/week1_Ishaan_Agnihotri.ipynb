### 4.3 Eigenvalues & Eigenvectors

For the matrix:
```

     [2, 3]]
```

1. Compute eigenvalues and eigenvectors using `np.linalg.eig`.
2. **Verify** the eigenvector equation: `A @ v = λ * v` for each eigenpair.
3. Plot both eigenvectors as arrows, scaled by their eigenvalues, on a 2D plot.
4. **Explain** in a markdown cell (below): what does it geometrically mean for a matrix to stretch a vector?

````python
A = np.array([[4, 1], [2, 3]], dtype=float)

eigenvalues, eigenvectors = np.linalg.eig(A)

print(f"Eigenvalues: {eigenvalues}")
print(f"Eigenvectors (columns):\n{eigenvectors}")

for i in range(len(eigenvalues)):
    lam = eigenvalues[i]
    v = eigenvectors[:, i]
    assert np.allclose(A @ v, lam * v)

plt.figure(figsize=(6, 6))

for i in range(len(eigenvalues)):
    lam = eigenvalues[i]
    v = eigenvectors[:, i]
    plt.quiver(0, 0, lam * v[0], lam * v[1],
               angles='xy', scale_units='xy', scale=1,
               label=f'λ={lam:.2f}')

plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.axhline(0, color='k', lw=0.8)
plt.axvline(0, color='k', lw=0.8)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Eigenvectors Scaled by Eigenvalues")
plt.legend()
plt.grid(True)

plt.show()

print("4.3 passed ✓")
````

### Geometric Explanation

Eigenvectors represent the directions that remain unchanged when a matrix transformation is applied. The matrix only changes their length, not their direction.

Eigenvalues tell us how much the corresponding eigenvectors are scaled:
- If the eigenvalue is greater than 1, the vector is **stretched**.
- If it is between 0 and 1, the vector is **compressed (squished)**.
- If it is negative, the vector is **flipped** and scaled.
- If it is 0, the vector collapses to the origin.

In this example, the matrix has eigenvalues **5** and **2**. This means one eigenvector is stretched by a factor of **5**, while the other is stretched by a factor of **2**. Since both eigenvalues are positive, the directions of the eigenvectors remain the same—they are only scaled. Geometrically, the transformation stretches space much more along the first eigenvector than along the second.

### 4.4 SVD & Dimensionality Reduction

1. Create a random 4×3 matrix `X` (use `np.random.seed(42)`).
2. Compute its **SVD**: `U, S, Vt = np.linalg.svd(X, full_matrices=False)`.
3. **Reconstruct** `X` from the SVD and verify it matches the original.
4. Perform a **rank-1 approximation** using only the largest singular value/vector — call it `X_approx`.
5. Explain the relationship between SVD components and PCA in the markdown cell below.

````python
np.random.seed(42)
X = np.random.randn(4, 3)

U, S, Vt = np.linalg.svd(X, full_matrices=False)

X_reconstructed = U @ np.diag(S) @ Vt

X_approx = np.outer(U[:, 0], Vt[0, :]) * S[0]

print(f"U shape: {U.shape}, S shape: {S.shape}, Vt shape: {Vt.shape}")
print(f"Singular values: {S}")
print(f"Reconstruction error: {np.linalg.norm(X - X_reconstructed):.2e}")
print(f"Rank-1 approximation:\n{X_approx}")

assert np.allclose(X, X_reconstructed)
print("4.4 passed ✓")
````

### SVD → PCA Connection

Singular Value Decomposition (SVD) decomposes a matrix **X** into three matrices:

- **U**: Contains the left singular vectors and represents the transformed data.
- **S**: Contains the singular values, which indicate the importance (amount of variance) captured by each component.
- **Vᵀ**: Contains the right singular vectors, which represent the principal directions (principal components).

PCA can be computed using SVD on a mean-centered dataset. In PCA:
- The **columns of V** (or rows of **Vᵀ**) are the **principal components**, i.e., the directions of maximum variance.
- The **singular values (S)** indicate how much variance each principal component explains. Larger singular values correspond to more important components.
- Keeping only the largest singular values and their corresponding vectors produces a lower-dimensional approximation of the data while preserving most of the information.

Therefore, the **Vᵀ matrix contains the directions of maximum variance**, and the **S matrix tells how much variance is captured along each of those directions.**

### 5.4 Distribution Testing & Stationarity

1. Generate two samples: `s1` ~ Normal(0, 1), `s2` ~ Exponential(1), both n=200 (seed=0).
2. Run the **Kolmogorov-Smirnov test** (`stats.kstest`) comparing each against a standard normal CDF.
3. Generate a non-stationary time series (linear trend + noise). Run the **Augmented Dickey-Fuller test** (`adfuller` from `statsmodels`) and interpret the result.
4. Difference the series once and re-run ADF to confirm stationarity.

````python
from scipy import stats
from statsmodels.tsa.stattools import adfuller
import numpy as np

np.random.seed(0)

s1 = np.random.normal(0, 1, 200)
s2 = np.random.exponential(1, 200)

ks_s1 = stats.kstest(s1, 'norm')
ks_s2 = stats.kstest(s2, 'norm')

print(f"KS test s1 (normal): stat={ks_s1.statistic:.4f}, p={ks_s1.pvalue:.4f}")
print(f"KS test s2 (exponential): stat={ks_s2.statistic:.4f}, p={ks_s2.pvalue:.4f}")

t = np.arange(200)
ts = 0.05 * t + np.random.normal(0, 1, 200)

adf_result = adfuller(ts)

print(f"\nADF statistic: {adf_result[0]:.4f}")
print(f"ADF p-value: {adf_result[1]:.4f}")
print(f"Stationary? {adf_result[1] < 0.05}")

ts_diff = np.diff(ts)

adf_diff = adfuller(ts_diff)

print(f"\nAfter differencing - p-value: {adf_diff[1]:.4f}, Stationary? {adf_diff[1] < 0.05}")
````

### 5.5 Model Monitoring Concepts

1. Implement `compute_psi(expected, actual, bins=10)` that computes the **Population Stability Index** between two distributions.
2. Simulate a scenario where a model's input distribution shifts and plot both distributions.
3. In the markdown cell, define: *concept drift*, *covariate drift*, PSI thresholds (what values indicate no/minor/major shift), and one trigger for retraining.

````python
def compute_psi(expected, actual, bins=10):
    eps = 1e-10

    breakpoints = np.linspace(np.min(expected), np.max(expected), bins + 1)

    expected_counts, _ = np.histogram(expected, bins=breakpoints)
    actual_counts, _ = np.histogram(actual, bins=breakpoints)

    expected_pct = expected_counts / len(expected) + eps
    actual_pct = actual_counts / len(actual) + eps

    psi = np.sum((actual_pct - expected_pct) *
                 np.log(actual_pct / expected_pct))

    return psi
````

### Concept Drift vs Covariate Drift & PSI Thresholds (5.5)

- **Concept Drift:** The relationship between input features and the target changes over time, causing the model's predictions to become less accurate.
- **Covariate Drift:** The distribution of the input features changes over time, while the relationship between inputs and the target remains the same.
- **PSI < 0.1:** No significant distribution shift; the model is considered stable.
- **PSI 0.1–0.2:** Moderate shift; monitor the model and investigate possible causes.
- **PSI > 0.2:** Significant distribution shift; retraining or updating the model is recommended.
- **Retraining trigger example:** Retrain the model when the PSI exceeds **0.2** or when model performance metrics (such as accuracy or F1-score) show a noticeable decline due to data drift.

### 6.1 Core Concepts

A bag contains: 4 red, 3 blue, 3 green marbles.

1. Define the sample space and compute P(red), P(blue), P(green).
2. Two marbles drawn **without replacement**. Compute:
   - P(first=red, second=blue) — **joint probability**
   - P(second=blue | first=red) — **conditional probability**
3. Are the two draws **independent**? Show mathematically.

````python
total = 10
red, blue, green = 4, 3, 3

p_red = red / total
p_blue = blue / total
p_green = green / total

p_red_then_blue = (red / total) * (blue / (total - 1))

p_blue_given_red = blue / (total - 1)

independent = (p_blue_given_red == p_blue)

print(f"P(red)={p_red:.3f}, P(blue)={p_blue:.3f}, P(green)={p_green:.3f}")
print(f"P(red,blue): {p_red_then_blue:.4f}")
print(f"P(blue|red): {p_blue_given_red:.4f}")
print(f"P(blue): {p_blue:.4f}")
print(f"Independent? {independent}")
````

### Probability Concepts (6.1)

- **Sample Space:** All possible marbles that can be drawn from the bag: 4 red, 3 blue, and 3 green (10 marbles total).

- **P(red):** \( \frac{4}{10} = 0.40 \)

- **P(blue):** \( \frac{3}{10} = 0.30 \)

- **P(green):** \( \frac{3}{10} = 0.30 \)

- **Joint Probability (First = Red, Second = Blue):**
  \[
  P(R \cap B) = P(R) \times P(B|R)
  = \frac{4}{10} \times \frac{3}{9}
  = \frac{2}{15} \approx 0.1333
  \]

- **Conditional Probability (Second = Blue | First = Red):**
  \[
  P(B|R) = \frac{3}{9} = \frac{1}{3} \approx 0.3333
  \]

- **Independence Check:**
  The two draws are **not independent** because
  \[
  P(B|R) = \frac{3}{9} = 0.3333 \neq P(B) = \frac{3}{10} = 0.30.
  \]
  Since the conditional probability is different from the original probability, drawing the first marble changes the probability of the second draw.

### 6.3 Bayes' Theorem

**Scenario — Spam filter:**
- P(Spam) = 0.30 (prior)
- P("free" | Spam) = 0.80 (likelihood)
- P("free" | Not Spam) = 0.05

1. Compute P(Spam | "free") using Bayes' theorem.
2. Implement a simple `naive_bayes_predict(prior_spam, p_word_given_spam, p_word_given_ham)` that returns the posterior P(Spam | word).
3. In the markdown cell, map each term — **prior, likelihood, evidence, posterior** — to the spam example.

````python
p_spam = 0.30
p_free_given_spam = 0.80
p_free_given_ham = 0.05
p_ham = 1 - p_spam

p_free = p_free_given_spam * p_spam + p_free_given_ham * p_ham

p_spam_given_free = (p_free_given_spam * p_spam) / p_free

print(f"P('free'): {p_free:.4f}")
print(f"P(Spam | 'free'): {p_spam_given_free:.4f}")

def naive_bayes_predict(prior_spam, p_word_given_spam, p_word_given_ham):
    p_ham = 1 - prior_spam
    p_word = p_word_given_spam * prior_spam + p_word_given_ham * p_ham
    return (p_word_given_spam * prior_spam) / p_word

pred = naive_bayes_predict(0.30, 0.80, 0.05)
assert abs(pred - p_spam_given_free) < 1e-9
print(f"naive_bayes_predict: {pred:.4f}")
print("6.3 passed ✓")
````

### Bayes' Theorem Terms (6.3)

- **Prior:** The initial probability that an email is spam before seeing any words.
  - **P(Spam) = 0.30**

- **Likelihood:** The probability of observing the word **"free"** given that the email is spam.
  - **P("free" | Spam) = 0.80**

- **Evidence:** The overall probability of observing the word **"free"** in any email.
  - **P("free") = P("free"|Spam) × P(Spam) + P("free"|Not Spam) × P(Not Spam)**

- **Posterior:** The updated probability that an email is spam after observing the word **"free"**.
  - **P(Spam | "free") ≈ 0.8727**

**Interpretation:** After seeing the word **"free"**, the probability that the email is spam increases from **30%** (prior) to about **87.27%** (posterior). This is the key idea behind Naive Bayes spam filtering.

| **Term**       | **Formula**    | **In this example**           |
| -------------- | -------------- | ----------------------------- |
| **Prior**      | P(Spam)        | **0.30**                      |
| **Likelihood** | P(word | Spam) | **P("free" | Spam) = 0.80**   |
| **Evidence**   | P(word)        | **P("free") = 0.275**         |
| **Posterior**  | P(Spam | word) | **P(Spam | "free") = 0.8727** |


### 6.4 Central Limit Theorem

1. Define a population following an **exponential distribution** (λ=1, size=100,000).
2. Repeatedly draw samples of size `n=30` and record their means — do this 5,000 times.
3. Plot the histogram of sample means.
4. Overlay the **theoretical normal distribution** predicted by the CLT: μ = population mean, σ = population_std / sqrt(n).
5. Run a KS test to confirm the distribution of sample means is approximately normal.

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

np.random.seed(7)
population = np.random.exponential(scale=1.0, size=100_000)

n_samples = 5000
sample_size = 30

sample_means = np.array([
    np.random.choice(population, sample_size).mean()
    for _ in range(n_samples)
])

pop_mean = population.mean()
pop_std = population.std()

clt_std = pop_std / np.sqrt(sample_size)

print(f"Population mean: {pop_mean:.4f}, Population std: {pop_std:.4f}")
print(f"Sample means mean: {sample_means.mean():.4f}, std: {sample_means.std():.4f}")
print(f"CLT predicted std: {clt_std:.4f}")

fig, ax = plt.subplots()

ax.hist(sample_means, bins=40, density=True, alpha=0.7, label='Sample Means')

x = np.linspace(sample_means.min(), sample_means.max(), 500)
ax.plot(
    x,
    stats.norm.pdf(x, pop_mean, clt_std),
    linewidth=2,
    label='CLT Normal'
)

ax.legend()
ax.set_title('CLT: Distribution of Sample Means')

plt.tight_layout()
plt.show()

ks_result = stats.kstest(sample_means, 'norm', args=(pop_mean, clt_std))
print(f"KS test p-value: {ks_result.pvalue:.4f} → Approximately normal? {ks_result.pvalue > 0.05}")
````


The Central Limit Theorem shows that when we take many samples, their averages tend to form a normal distribution, even if the original data is not normal. This helps us make reliable conclusions from sample data and forms the basis of many statistical tests and confidence intervals.
